```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

State-Wise Distribution of EV Categories

```python
df = pd.read_csv('EV Stat.csv')
# print(df.head(15))

# Plotting a grouped bar chart
df.plot(x='State', kind='bar', stacked=False, figsize=(9, 4))
plt.title('Distribution of EV Categories Across States')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90)
plt.show()
```

State-Wise Total EV Count

```python
pivot_table = df.pivot_table(index="State", values="Total in state")
plt.figure(figsize=(9, 8))
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", linewidths=.5)
plt.title('Total EV Count by State')
plt.show()
```

Comparison of Two-Wheeler Categories Across States

```python
state = "Maharashtra"  # Example state
state_data = df[df['State'] == state].drop(columns=['State', 'Total in state']).sum()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(state_data, labels=state_data.index, autopct='%1.1f%%', startangle=140)
plt.title(f'Category Contribution to Total EVs in {state}')
plt.show()
```

Comparison of Two-Wheeler Categories Across States

```python
# Selecting relevant columns
two_wheeler_columns = ['Two Wheelers (Category L1 & L2 as per CMVR)',
                       'Two Wheelers (Category L2 (CMVR))',
                       'Two Wheelers (Max power not exceeding 250 Watts)']

df[two_wheeler_columns + ['State']].set_index('State').plot(kind='bar', stacked=True, figsize=(10, 8))
plt.title('Comparison of Two-Wheeler Categories Across States')
plt.ylabel('Number of Vehicles')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(df.drop(columns=['State', 'Total in state']))
plt.title('Correlation Between Vehicle Categories')
plt.show()
```
